```python
def split_variables(file_path):
    ''' Divide the data into
        Statistical features of explanatory variables and
        traditional market variables'''
    total = pd.read_csv(file_path)
    total.fillna(method='ffill', inplace=True)
    # Convert the 'Date' column to datetime
    total['Date'] = pd.to_datetime(total['Date'])
    total.set_index(total['Date'], inplace=True)
    total.drop(['Date','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'],axis = 1,inplace = True)
    variables = total.iloc[:,:25]
    traditional_market = total.iloc[:,25:]
    return total,variables, traditional_market
data,variables,traditional_market = split_variables("jrfm-2156907.csv")

def table2(var):
    return var.describe().T[['count','mean','std','min','max']]
des_var = table2(variables)
des_tra = table2(traditional_market)
des_var
```

```python

def prepare_datasets(file_path, start_day, test_day, end_day = date(2022,4,1)):
    # Load and preprocess data
    total = pd.read_csv(file_path)
    total['Date'] = pd.to_datetime(total['Date']).dt.date
    total = total[(total['Date'] >= start_day) & (total['Date'] <= end_day)]
    total.fillna(method='ffill', inplace=True)

    # Split into train and test datasets
    train_dataset = total[total['Date'] < test_day].reset_index(drop=True)
    test_dataset = total[total['Date'] >= test_day].reset_index(drop=True)

    # Separate features and target variable
    X_train = train_dataset.drop(['Date', 'BTC_Close'], axis=1)[:-1]
    Y_train = train_dataset['BTC_Close'][1:]

    X_test = test_dataset.drop(['Date', 'BTC_Close'], axis=1)[:-1]
    Y_test = test_dataset['BTC_Close'][1:]

    return X_train, Y_train, X_test, Y_test

# Usage
Start_day_o1 = date(2015, 8, 7)
Test_day_o1 = date(2018, 4, 1)
End_day_o1 = date(2018, 9, 30)
Start_day_o2 = date(2018, 10, 1)
Test_day_o2 = date(2021, 10, 1)

X_train_o1, Y_train_o1, X_test_o1, Y_test_o1 = prepare_datasets("jrfm-2156907.csv", Start_day_o1, Test_day_o1, End_day_o1)
X_train_o2, Y_train_o2, X_test_o2, Y_test_o2 = prepare_datasets("jrfm-2156907.csv", Start_day_o2, Test_day_o2)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

def plot_bitcoin_price_periods(csv_file_path):
    # Load the dataset
    data = pd.read_csv(csv_file_path)

    # Ensure that 'Date' column is a datetime type
    data['Date'] = pd.to_datetime(data['Date'])

    # Set date as the index
    data.set_index('Date', inplace=True)

    # Define the periods based on the provided input
    periods = {
        'Period 1 Train': ('2015-03-31', '2018-03-31'),
        'Period 1 Test': ('2018-04-01', '2018-09-30'),
        'Period 2 Train': ('2018-10-01', '2021-09-30'),
        'Period 2 Test': ('2021-10-01', '2022-04-01')
    }

    # Initialize the plot
    plt.figure(figsize=(15, 7))

    # Plot Bitcoin Price
    plt.plot(data.index, data['BTC_Close'], label='Bitcoin Price', color='red')

    # Highlight the periods on the plot
    for period, (start, end) in periods.items():
        plt.axvspan(datetime.strptime(start, '%Y-%m-%d'),
                    datetime.strptime(end, '%Y-%m-%d'),
                    label=period.replace('Period', 'P'),
                    color='yellow' if 'Train' in period else 'green',
                    alpha=0.3)

    # Add labels and title
    plt.xlabel('Date')
    plt.ylabel('Bitcoin Price')
    plt.title('Interval division of training samples and test samples')
    plt.legend()

    # Save the figure
    plt.tight_layout()
    figure_path = 'annotated_bitcoin_price_periods.png'
    plt.savefig(figure_path)

    # Close the plot
    plt.close()

    # Return the path to the saved figure
    return figure_path

# Usage
csv_file_path = 'jrfm-2156907.csv'  # Replace with the actual path to your CSV file
saved_figure_path = plot_bitcoin_price_periods(csv_file_path)
print(f"The figure has been saved to: {saved_figure_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

def plot_bitcoin_price_periods(csv_file_path, save_file_path):
    # Load the dataset
    data = pd.read_csv(csv_file_path)

    # Ensure that 'Date' column is a datetime type
    data['Date'] = pd.to_datetime(data['Date'])

    # Set date as the index
    data.set_index('Date', inplace=True)

    # Define the periods with train and test splits
    periods = {
        'Period 1 Train': (Start_day_n1, Test_day_n1),
        'Period 1 Test': (Test_day_n1, End_day_n1),
        'Period 2 Train': (Start_day_n2, Test_day_n2),
        'Period 2 Test': (Test_day_n2, End_day_n2),
        'Period 3 Train': (Start_day_n3, Test_day_n3),
        'Period 3 Test': (Test_day_n3, End_day_n3),
        'Period 4 Train': (Start_day_n4, Test_day_n4),
        'Period 4 Test': (Test_day_n4, End_day_n4),

    }

    # Initialize the plot
    plt.figure(figsize=(15, 7))

    # Plot Bitcoin Price
    plt.plot(data.index, data['BTC_Close'], label='Bitcoin Price', color='red')

    # Define colors for the training and testing periods
    train_color = 'yellow'
    test_color = 'green'

    # Highlight the periods on the plot
    for period, (start, end) in periods.items():
        plt.axvspan(start,
                    end,
                    label=period,
                    color=train_color if 'Train' in period else test_color,
                    alpha=0.3)

    # Add labels and title
    plt.xlabel('Date')
    plt.ylabel('Bitcoin Price')
    plt.title('Bitcoin Price with Highlighted Training and Testing Periods')
    plt.legend()

    # Save the figure
    plt.tight_layout()
    plt.savefig(save_file_path)

    # Close the plot to avoid display overhead
    plt.close()

    # Return the path to the saved figure
    return save_file_path

# Specify the file paths
csv_file_path = 'jrfm-2156907.csv'  # CSV file path
save_file_path = 'annotated_bitcoin_price_train_test_periods_new_periods.png'  # Path to save the figure

# Call the function and save the path to the figure
figure_path = plot_bitcoin_price_periods(csv_file_path, save_file_path)
print(f"The figure has been saved to: {figure_path}")
```
